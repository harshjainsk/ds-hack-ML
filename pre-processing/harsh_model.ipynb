# converting symptoms to numbers

```python
new_dict = dict([(value, key) for key, value in symp_dict.items()])
new_dict['dischromic _patches']=102
symp_dict
```

```python
def convert_symptom_to_number(dataframe, column_name):
    
    dataframe[column_name]= dataframe[column_name].str.strip()
    dataframe[column_name]= dataframe[column_name].map(new_dict)
```

```python
convert_symptom_to_number(symptoms, 'Symptom_1')
convert_symptom_to_number(symptoms, 'Symptom_2')
convert_symptom_to_number(symptoms, 'Symptom_3')
```

# Preprocessing the disease data

```python
dis_dict= disease['Disease'].to_dict()
dis_dict
```

```python
temp = []
res_dict = dict()
for key, val in dis_dict.items():
    if val not in temp:
        temp.append(val)
        res_dict[key] = val
```

```python
dis_dict
```

```python
new_res_dict = dict([(value, key) for key, value in res_dict.items()])
new_res_dict
```

# converting the preprocessed disease data to a csv

```python
disease['Disease'] = disease['Disease'].map(new_res_dict)
disease.to_csv('disease_preprocessed_data', index=False)
```
